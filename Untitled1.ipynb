```python
class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def eje_x(self):
        return self.x

    def eje_y(self):
        return self.y

    def impresion(self):
        return "El punto se encuentra en x = " + str(self.x) + " y en y = " + str(self.y)

    def opuesto(self):
        return Punto(-self.x, -self.y)
```

```python
class Linea:
  def __init__(self, punto1, punto2):
    self.p1 = punto1
    self.p2 = punto2

  def mueveDerecha(self, n):
    self.p1.x += n
    self.p2.x += n

  def mueveIzquierda(self, n):
    self.p1.x -= n
    self.p2.x -= n

  def mueveArriba(self, n):
    self.p1.y += n
    self.p2.y += n

  def mueveAbajo(self, n):
    self.p1.y -= n
    self.p2.y -= n
```
